```python
import matplotlib.pyplot as plt
import pylab
import numpy as np
import time
import math
import csv
```

```python
#Question1
seedInt = time.time()
seedInt = int(1000 * (seedInt - math.floor(seedInt)))
def randomInt(min, max):
    global seedInt
    seedInt = ((seedInt + 47) % (max - min)) + min
    return seedInt
```

```python
#Question1
def randomFloat(min, max):
    seedFloat = time.time()
    seedFloat = 1000 * (seedFloat - math.floor(seedFloat))
    return ((seedFloat + 47) % (max - min)) + min
```

```python
#Question2
arrival_times = []
inter_arrival_times = []
analysis_data = []
validation_data = []

with open('customer_arrivals.csv', 'r') as csvFile:
    myReader = csv.reader(csvFile)
    for row in myReader:
        arrival_times.append(float((row[0].strip())))
        
csvFile.close()

inter_arrival_times.append(arrival_times[0]) 
for i in range(1,len(arrival_times)):
    inter_arrival_times.append(round(arrival_times[i] - arrival_times[i - 1], 2))

for i in range(0,int(len(inter_arrival_times)/10)):
    validation_data.append(inter_arrival_times[i])

for i in range(int(len(inter_arrival_times)/10),len(inter_arrival_times)):
    analysis_data.append(inter_arrival_times[i])
    
inter_arrival_times.sort()
```

```python
#Question2
#Arrival Times
print(arrival_times)
```

```python
#Question2
#Inter Arrival Times Sorted
print(inter_arrival_times)
```

```python
#Question2
#Sorted Inter Arrival Times Histogram
plt.hist(inter_arrival_times,bins=75)
plt.title('Histogram of Sorted Inter Arrival Times')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.show()
```

```python
#Question2
#Analysis Data Histogram
plt.hist(analysis_data,bins=75)
plt.title('Histogram of Analysis Data')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.show()
```

```python
#Question2
#Validation Data Histogram
plt.hist(validation_data,bins=75)
plt.title('Histogram of Validation Data')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.show()
```
