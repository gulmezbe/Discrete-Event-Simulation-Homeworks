```python
from random import *
import random
import matplotlib.pyplot as plt
import pylab
import numpy as np
```

```python
#Part 1.1
def gender():
    return choice(['Male', 'Female'])
```

```python
#Part 1.2
def num_females(n):
    females = 0
    for x in range(0, n):
        if gender() == 'Female':
            females = females + 1
    return females
```

```python
#Part 1.2
n = 20
female = num_females(n)
male = n - female
thisdict = {
  "Males": male / n,
  "Females": female / n
}
print(thisdict)
plt.bar(thisdict.keys(), thisdict.values(),width = 0.6, color='g')
```

```python
#Part 1.3
def female_height():
    return gauss(158.9, 0.6)
```

```python
#Part 1.3
def male_height():
    return gauss(174.1, 0.7)
```

```python
#Part 1.3
xx = []
for i in range(0,20):
    xx.append(female_height())
xx = np.asarray(xx)
n, bins, patches = plt.hist(xx, 'auto', density=True, facecolor='g', alpha=0.75)
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.title('Height of Females')
plt.axis([156, 162, 0, 2])
plt.grid(True)
plt.show()
```

```python
#Part 1.3
xy = []
for i in range(0,20):
    xy.append(male_height())
xy = np.asarray(xy)
n, bins, patches = plt.hist(xy, 'auto', density=True, facecolor='g', alpha=0.75)
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.title('Height of Males')
plt.axis([170, 178, 0, 2])
plt.grid(True)
plt.show()
```

```python
#Part 1.4
def expected_height_by_gender(gender, n):
    if gender == 'Male':
        male_total = 0
        for i in range(0, n):
            male_total = male_total + male_height()
        return male_total/n
    if gender == 'Female':
        female_total = 0
        for i in range(0, n):
            female_total = female_total + female_height()
        return female_total/n
```

```python
#Part 1.4
expected_height = []
for i in range(0,20):
    expected_height.append(expected_height_by_gender('Female', 20))
expected_height = np.asarray(expected_height)
n, bins, patches = plt.hist(expected_height, 'auto', density=True, facecolor='g', alpha=0.75)
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.title('Expected Height of Females')
plt.axis([156, 162, 0, 5])
plt.grid(True)
plt.show()  

xx = []
for i in range(0,20):
    xx.append(female_height())
xx = np.asarray(xx)
n, bins, patches = plt.hist(xx, 'auto', density=True, facecolor='g', alpha=0.75)
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.title('Height of Females')
plt.axis([156, 162, 0, 2])
plt.grid(True)
plt.show()
```

```python
#Part 1.4
#Yes there is a different in the two diagrams first diagram gave close results because first diagram takes the average of results.
#If we reran experiment or if we use increase run time of function result will be similar.
```

```python
#Part 1.5
def expected_height(n):
    total = 0
    for i in range(0,n):
        if gender() == 'Male':
            total = total + male_height()
        else:
            total = total + female_height()
    return total / n
```

```python
#Part 1.5
array = []
for i in range(0,10000):
    array.append(expected_height(1))
array = np.asarray(array)
n, bins, patches = plt.hist(array, 'auto', density=True, facecolor='g', alpha=0.75)
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.title('Height')
plt.axis([156, 177, 0, 0.5])
plt.grid(True)
plt.show()
```

```python
#Part 2.1
def inter_arrival_time():
    return round(random.uniform(0, 3), 2)
```

```python
#Part 2.2
arrival = []
for i in range(0,1000):
    arrival.append(inter_arrival_time())
arrival = np.asarray(arrival)
n, bins, patches = plt.hist(arrival, 'auto', density=True, facecolor='g', alpha=0.75)
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.title('Inter Arrival Time')
plt.axis([0, 3, 0, 1])
plt.grid(True)
plt.show()
```

```python
#Part 2.3
def delivery_time():
    return round(random.uniform(1, 15), 2)
```

```python
#Part 2.3
def rot_time():
    return round(random.uniform(7, 12), 2)
```

```python
#Part 2.3
Delivery = []
for i in range(0,1000):
    Delivery.append(delivery_time())
Delivery = np.asarray(Delivery)
n, bins, patches = plt.hist(Delivery, 'auto', density=True, facecolor='g', alpha=0.75)
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.title('Delivery Time')
plt.axis([1, 15, 0, 1])
plt.grid(True)
plt.show()
```

```python
#Part 2.3
rot = []
for i in range(0,1000):
    rot.append(rot_time())
rot = np.asarray(rot)
n, bins, patches = plt.hist(rot, 'auto', density=True, facecolor='g', alpha=0.75)
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.title('Rot Time')
plt.axis([7, 12, 0, 1])
plt.grid(True)
plt.show()
```

```python
#Part 2.4
def number_of_cabbages():
    return randint(0, 40)
```

```python
#Part 2.4
cabbages = number_of_cabbages()
need = 40 - cabbages
day = 0
print("Day = ", round(day, 2), "    Cabbages = ", cabbages)
for i in range(0,need):
    time = delivery_time()
    print("Cabbage ordered, delivery time = ", time)
    day += time
    cabbages += 1
    print("Day = ", round(day, 2), "    Cabbages = ", cabbages)
```
